```python
import numpy as np
```

```python
temperaturas = np.array([30,40,50, 263, 78, 52, 53])
print(temperaturas)
print(type(temperaturas))
```

```python
media = np.mean(temperaturas)
print(f'{media:.2f}')
```

```python
maior_valor = np.max(temperaturas)
print(maior_valor)
```

```python
mediana = np.median(temperaturas)
print(mediana)
```

```python
cidades = [
    "São Paulo",
    "Campinas",
    "Santos",
    "Ribeirão Preto",
    "Sorocaba"
]

habitantes = [
    11451999,  # São Paulo
    1187024,   # Campinas
    418608,    # Santos
    728400,    # Ribeirão Preto
    757459     # Sorocaba
]
```

```python
numero_habitantes = np.array(habitantes)

maior_populacao = np.max(numero_habitantes)
print(f'{maior_populacao:.2f}')

menor_populacao = np.min(numero_habitantes)
print(f'{menor_populacao}')

media_populacao = np.mean(numero_habitantes)
print(f'{media_populacao}')

mediana_populacao = np.median(numero_habitantes)
print(f'{mediana_populacao}')

```
