```python
import pandas as pd
```

```python
##Abrir arquivo txt.

arquivo = open(r'data\dom_casmurro.txt', encoding = 'utf-8')
```

```python
#Ler arquivo txt.

texto = arquivo.read()

len(texto)
```

```python
#Transformar o arquivo em uma lista

palavras = texto.split()

len(palavras)

palavras
```

```python
#Criar um dataframe da lista criada

tabela = pd.DataFrame({'palavras': palavras})
tabela
```

```python
#Classificar as palavras mais usadas

palavras_mais_usadas = tabela['palavras'].value_counts()

palavras_mais_usadas.head(20)
```

```python
#Retirar a ultima palavra da lista criando outra coluna, assim entendemos como funcionado uma IA generativa

tabela['proxima_palavra'] = tabela['palavras'].shift(-1)

tabela
```

```python
#Retirar valores nulos 

tabela = tabela.dropna()

tabela
```

```python
#Selecionar uma palavra do txt

filtro = tabela['palavras'] == 'feliz'
```

```python
#Mostra quantas vezes aparece no texto

tabela[filtro]['palavras'].value_counts()
```

```python
# Conta quantas vezes a proxima palavra aparece a partir da palavra escolhida

tabela[filtro]['proxima_palavra'].value_counts()
```

```python
#Probabilidade da próxima palavra aparecer

tabela[filtro]['proxima_palavra'].value_counts(normalize=True)
```

```python
#Filtrar a palavra com maior probabilidade de ser a proxima

tabela[filtro]['proxima_palavra'].value_counts().idxmax()
```
