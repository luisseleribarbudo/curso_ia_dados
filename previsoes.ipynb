```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'.\data/dataset_06072026.csv')
```

```python
sns.lmplot(data=df, x='study_hours_per_day', y='exam_score');
```

```python
X = df[['study_hours_per_day',
    'age',
    'social_media_hours',
    'netflix_hours',
    'sleep_hours',
    'exercise_frequency',
    'mental_health_rating'
]]
y = df['exam_score']

from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X,y)
```

```python
novo_aluno = pd.DataFrame({'study_hours_per_day': [1]})
modelo.predict(novo_aluno)
```

```python
modelo.score(X,y)
```
