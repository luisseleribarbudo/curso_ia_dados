```python
avaliacao = float(input('Qual foi sua nota?'))

while True:

    if avaliacao > 8:
        print('Excelente!!')
    elif avaliacao > 6:
        print('Nota ok!')
    elif avaliacao > 4:
        print('Estude mais!')
    else:
        print('Procure outra coisa para fazer')
```
