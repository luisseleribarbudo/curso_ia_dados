```python
import pandas as pd
import numpy as np
```

```python
quantidade_produtos = [3,5,6,10]
precos_produtos = [10,5,7,100]

## criar arrays para que possamos manipular os dados

array_quantidade = np.array(quantidade_produtos)
array_precos = np.array(precos_produtos)

faturamento = array_quantidade * array_precos
print(faturamento)
```

```python
quantidade_produtos = [3,5,6,10]
precos_produtos = [10,5,7,100]
nomes_produtos = ['pastel', 'caçulinha', 'refrigerante', 'combo']

## criar arrays para que possamos manipular os dados

array_quantidade = np.array(quantidade_produtos)
array_precos = np.array(precos_produtos)

faturamento = array_quantidade * array_precos
for nome, preco in zip(nomes_produtos, faturamento):
    print(f'{nome} : R${preco}')
```

```python
# Lista com o nome de 10 produtos
produtos = [
    "Arroz",
    "Feijão",
    "Macarrão",
    "Óleo",
    "Açúcar",
    "Café",
    "Leite",
    "Sabão em Pó",
    "Detergente",
    "Papel Higiênico"
]

# Lista com os preços de custo
preco_custo = [
    18.50,
    7.20,
    4.10,
    5.80,
    3.90,
    12.50,
    4.80,
    10.00,
    2.30,
    15.00
]

# Lista com os preços de venda
preco_venda = [
    24.90,
    9.50,
    5.90,
    7.90,
    5.50,
    16.90,
    6.50,
    14.90,
    3.90,
    19.90
]

array_produtos = np.array(produtos)
array_custo = np.array(preco_custo)
array_venda = np.array(preco_venda)

lucro = array_venda - array_custo
for nome, valor, custeio in zip(produtos, lucro, preco_custo):
    print(f'O lucro do {nome} foi de R${valor:.2f} e o produto custou R${custeio}')
    
lucro_total = np.sum(lucro)
print(f'O lucro total é de R$ {lucro_total}')
```
